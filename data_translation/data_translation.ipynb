```python
translator = Translator()
```

increase the timeout duration to prevent timeout error

```python
def translate_text(text):
    return translator.translate(text, dest='en').text

def translate_with_timeout(x):
    try:
        with concurrent.futures.ThreadPoolExecutor() as executor:
            future = executor.submit(translate_text, x)
            return future.result(timeout=60)  # Adjust the timeout duration as needed
    except concurrent.futures.TimeoutError:
        return "Translation timeout"  # or any other appropriate error message
```

# Replace non-readable values with an empty string


```python
content_translated_final['text'] = content_translated_final['text'].replace('[^\x00-\x7F]+', '', regex=True)
```

decrease text size to 800 

```python
def truncate_text(text, length=800):
    if len(text) > length:
        truncated_text = text[:length]
        return truncated_text
    else:
        return text
```

replace Nan values with an empty string

```python
content_translated_final['text'] = content_translated_final['text'].replace(np.nan, '')
```

```python
content_translated_final['text'] = content_translated_final['text'].apply(truncate_text)
```

```python
content_translated_final['text'] = content_translated_final['text'].swifter.apply(lambda x: translator.translate(x, dest='en', timeout = 380).text)
```
