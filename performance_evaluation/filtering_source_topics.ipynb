```python
import pandas as pd

topic = pd.read_csv("topics_doc2vec.csv")
correlation = pd.read_csv("correlations.csv") # correlations.csv can be uploaded from https://www.kaggle.com/competitions/learning-equality-curriculum-recommendations/data?select=correlations.csv
result_doc2vec = pd.read_csv("results_doc2vec_new_params (7).csv")
result_tf_idf = pd.read_csv("results_tf-idf_all-lang.csv")
```

```python
topic
```

filter source catergories

```python
topic_source = topic[topic['category']=='source']
topic_source
```

```python
ids_to_filter = topic_source['id']
ids_to_filter
```

```python
result_tf_idf_source = result_tf_idf[result_tf_idf['topic_id'].isin(ids_to_filter)]
result_doc2vec_source = result_doc2vec[result_doc2vec['topic_id'].isin(ids_to_filter)]
correlation_source = correlation[correlation['topic_id'].isin(ids_to_filter)]
```

save filtered result_tf_idf, result_doc2vec, correlation

```python
result_doc2vec_source.to_csv('results_doc2vec.csv', index=False)
result_tf_idf_source.to_csv('results_tf_idf.csv', index=False)
correlation_source.to_csv('ground_truth.csv', index=False)
```
